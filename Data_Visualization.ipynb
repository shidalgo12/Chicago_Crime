```python
# Pull Crime Data file
file_to_load = "City Data/area_crimes.csv"
# Read Crime Data File
area_crime = pd.read_csv(file_to_load)
area_crime

# Highlight Areas with the Max Amount of Crimes per Type
def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

area_crime.style.apply(highlight_max)
```
